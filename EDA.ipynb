**IMPORTING LIBRARIES**

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
import seaborn as sns
```

**READING THE DATASET**

```python
my_data = pd.read_csv('/Student Performance new.csv')
my_data.head(5)
```

```python
my_data.isnull()
```

```python
sns.lineplot(data=my_data.head(5),x="math percentage",y="reading score percentage")
plt.show()
```

**HEAT MAP-[NULL VALUES]**

```python
sns.heatmap(my_data.head(5).isnull())
```

**COUNT PLOT**

```python
sns.set_style('darkgrid')
sns.countplot(x="reading score percentage",hue="writing score percentage",data=my_data[25:28],palette='OrRd')
```

```python
sns.countplot(x="math percentage",data=my_data[28:32],palette="BuGn")
```

**DISPLOT**

```python
sns.displot(my_data['writing score percentage'].dropna(),color='black',bins=8)
```

**BAR PLOT**

```python
sns.barplot(data=my_data[28:34],x='parental level of education',y="writing score percentage")
```

**BOX-PLOT**

```python
plt.figure(figsize=(8,6))
sns.boxplot(x="math percentage",y="writing score percentage",data=my_data.head(5),palette='PiYG')
```

**SCATTER PLOT**

```python
plt.figure(figsize=(8,6))
sns.scatterplot(x="math percentage",y="reading score percentage",data=my_data,palette='Darkblue')
```

```python
sns.pairplot(my_data[0:51])
```

**CAT** **PLOT**

```python
sns.catplot(data=my_data[0:10],x="writing score percentage",y="reading score percentage")
```
